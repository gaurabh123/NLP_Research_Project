```python
import pandas as pd
import os
from sklearn.utils import shuffle

folder = "val"

allFiles = os.listdir(folder)

temp_list = []
for files in allFiles:
    temp_list.append(pd.read_csv("./{}/{}".format(folder,files), sep='\t'))

all_df = pd.concat(temp_list)
all_df = shuffle(all_df, random_state=7)
all_df.to_csv("./MergedData/{}DataMerged.tsv".format(folder), sep='\t')
```
